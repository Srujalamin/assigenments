7.Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings. 

```python
def match_words(words):
  ctr = 0

  for word in words:
    if len(word) > 1 and word[0] == word[-1]:
      ctr += 1
  return ctr

print(match_words(['abc', 'xyx', 'abc', '1221','ksk']))
```

52.Write a Python function to check whether a number is perfect or not

```python
def perfect_number(n):
    sum = 0
    for x in range(1, n):
        if n % x == 0:
            sum += x
    return sum == n
print(perfect_number(6))
```

53.Write a Python function that checks whether a passed string is palindrome or not 

```python
def isPalindrome(string):
	left_pos = 0
	right_pos = len(string) - 1
	
	while right_pos >= left_pos:
		if not string[left_pos] == string[right_pos]:
			return False
		left_pos += 1
		right_pos -= 1
	return True
print(isPalindrome('aza')) 
```

67.Write a Python program to returns sum of all divisors of a number

```python
def sum_div(number):
    divisors = [1]
    for i in range(2, number):
        if (number % i)==0:
            divisors.append(i)
    return sum(divisors)
print(sum_div(8))
print(sum_div(12))
```
